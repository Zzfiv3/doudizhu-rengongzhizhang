```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

kicker_accuracy = pd.read_csv('kicker_accuracy_df.csv')
kicker_loss = pd.read_csv('kicker_loss_df.csv')
policy_accuracy = pd.read_csv('policy_accuracy_df.csv')
policy_loss = pd.read_csv('policy_loss_df.csv')
```

```python
N = 1000
n = np.ones(N)
weights = n / N
sma = np.convolve(weights, kicker_loss.loss, mode='valid')
```

```python
# train and test accuracy of the Kicker Net
plt.figure(figsize=(16, 5))
plt.subplot(1, 2, 1)
plt.plot(kicker_accuracy.train_accuracy, '-o', lw=2, label='Train')
plt.plot(kicker_accuracy.test_accuracy, '-o', lw=2, label='Test')
plt.title('Training and test accuracy of the Kicker Net', size=18)
plt.xlabel('Number of epochs', size=16)
plt.ylabel('Accuracy', size=16)
plt.legend(fontsize=14)


plt.subplot(1, 2, 2)
plt.plot(sma, lw=2)
plt.title('Loss of the Kicker Net in the training process', size=18)
plt.xlabel('Number of batches', size=16)
plt.ylabel('Loss', size=16)

plt.savefig('kicker.png', dpi=300)
plt.show()
```
